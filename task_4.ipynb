```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Road Accident Data.csv.zip")
print(df.head())
```

```python
print("Shape of dataset:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())
```

```python
if "Time" in df.columns:
    df["Hour"] = pd.to_datetime(df["Time"], errors="coerce").dt.hour
```

```python
for col in ["Weather_Conditions", "Road_Surface_Conditions", "Urban_or_Rural_Area"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
```

```python
if "Weather_Conditions" in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x="Weather_Conditions", order=df["Weather_Conditions"].value_counts().index)
    plt.xticks(rotation=45)
    plt.title("Accidents by Weather Conditions")
    plt.show()
```

```python
if "Hour" in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df["Hour"].dropna(), bins=24, kde=False)
    plt.title("Accidents by Hour of Day")
    plt.xlabel("Hour (0 = Midnight)")
    plt.ylabel("Number of Accidents")
    plt.show()
```

```python
if "Road_Surface_Conditions" in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x="Road_Surface_Conditions", order=df["Road_Surface_Conditions"].value_counts().index)
    plt.xticks(rotation=45)
    plt.title("Accidents by Road Surface Conditions")
    plt.show()
```

```python
if "Urban_or_Rural_Area" in df.columns:
    plt.figure(figsize=(6,5))
    sns.countplot(data=df, x="Urban_or_Rural_Area")
    plt.title("Urban vs Rural Accidents")
    plt.show()
```
